```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid")
pd.set_option("display.max_columns", 200)
pd.set_option("display.max_colwidth", 120)

# -------- 0) Basic profile --------
print("Shape:", df1.shape)
print("\nData types:")
display(df1.dtypes.to_frame("dtype"))

numeric_cols = df1.select_dtypes(include=["number"]).columns.tolist()
cat_cols = df1.select_dtypes(exclude=["number"]).columns.tolist()

print(f"\nNumeric columns ({len(numeric_cols)}):", numeric_cols)
print(f"Categorical columns ({len(cat_cols)}):", cat_cols)

print("\nPreview:")
display(df1.head())

# -------- 1) Data quality audit --------
missing = df1.isna().sum().sort_values(ascending=False)
missing_pct = (missing / len(df1) * 100).round(2)
quality_report = pd.DataFrame({
    "missing_count": missing,
    "missing_pct": missing_pct,
    "n_unique": df1.nunique(dropna=False)
}).sort_values(["missing_count", "n_unique"], ascending=[False, False])

print("\nDuplicate rows:", df1.duplicated().sum())
print("\nQuality report (top 20):")
display(quality_report.head(20))

plt.figure(figsize=(10, 4))
sns.barplot(x=missing_pct.index, y=missing_pct.values, palette="viridis")
plt.title("Missing Values (%) by Column")
plt.xticks(rotation=45, ha="right")
plt.ylabel("Missing %")
plt.xlabel("Columns")
plt.tight_layout()
plt.show()

# -------- 2) Univariate analysis --------
if numeric_cols:
    n = len(numeric_cols)
    cols = 3
    rows = int(np.ceil(n / cols))

    fig, axes = plt.subplots(rows, cols, figsize=(5 * cols, 3.5 * rows))
    axes = np.array(axes).reshape(-1)

    for i, col in enumerate(numeric_cols):
        sns.histplot(df1[col].dropna(), kde=True, ax=axes[i], color="#4C72B0")
        axes[i].set_title(f"Distribution: {col}")

    for j in range(i + 1, len(axes)):
        axes[j].axis("off")

    plt.tight_layout()
    plt.show()

    # Boxplots for outlier visibility
    n = len(numeric_cols)
    cols = 3
    rows = int(np.ceil(n / cols))

    fig, axes = plt.subplots(rows, cols, figsize=(5 * cols, 3.5 * rows))
    axes = np.array(axes).reshape(-1)

    for i, col in enumerate(numeric_cols):
        sns.boxplot(x=df1[col], ax=axes[i], color="#55A868")
        axes[i].set_title(f"Boxplot: {col}")

    for j in range(i + 1, len(axes)):
        axes[j].axis("off")

    plt.tight_layout()
    plt.show()

if cat_cols:
    for col in cat_cols[:8]:  # keep readable
        plt.figure(figsize=(8, 4))
        order = df1[col].astype(str).value_counts().index
        sns.countplot(data=df1, x=col, order=order, palette="mako")
        plt.title(f"Category Frequency: {col}")
        plt.xticks(rotation=45, ha="right")
        plt.tight_layout()
        plt.show()

# -------- 3) Correlation and interactions --------
if len(numeric_cols) >= 2:
    corr = df1[numeric_cols].corr(numeric_only=True)
    plt.figure(figsize=(1.1 * len(numeric_cols) + 4, 0.9 * len(numeric_cols) + 3))
    sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", center=0, square=True)
    plt.title("Correlation Heatmap (Numeric Features)")
    plt.tight_layout()
    plt.show()

    # Top absolute correlations (excluding diagonal duplicates)
    corr_pairs = (
        corr.where(np.triu(np.ones(corr.shape), k=1).astype(bool))
        .stack()
        .reset_index()
    )
    corr_pairs.columns = ["feature_1", "feature_2", "correlation"]
    corr_pairs["abs_corr"] = corr_pairs["correlation"].abs()
    print("\nTop 10 absolute correlations:")
    display(corr_pairs.sort_values("abs_corr", ascending=False).head(10))

# -------- 4) Target-aware analysis (auto-detect) --------
target_candidates = [
    "Performance Index", "performance_index", "final_grade", "Final Grade", "G3",
    "math score", "reading score", "writing score", "Score", "score"
]
target_col = next((c for c in target_candidates if c in df1.columns), None)

if target_col is None and numeric_cols:
    # fallback: assume last numeric as potential target for quick exploratory insights
    target_col = numeric_cols[-1]

print("\nSelected target-like column for focused EDA:", target_col)

if target_col is not None and target_col in numeric_cols:
    # feature-vs-target scatter for strongest correlated features
    if len(numeric_cols) >= 2:
        corr_to_target = (
            df1[numeric_cols].corr(numeric_only=True)[target_col]
            .drop(target_col, errors="ignore")
            .abs()
            .sort_values(ascending=False)
        )
        top_feats = corr_to_target.head(min(4, len(corr_to_target))).index.tolist()

        for feat in top_feats:
            plt.figure(figsize=(6, 4))
            sns.regplot(data=df1, x=feat, y=target_col, scatter_kws={"alpha": 0.6}, line_kws={"color": "red"})
            plt.title(f"{feat} vs {target_col}")
            plt.tight_layout()
            plt.show()

    # categorical vs target distributions
    for col in cat_cols[:4]:
        plt.figure(figsize=(8, 4))
        sns.boxplot(data=df1, x=col, y=target_col, palette="Set2")
        plt.title(f"{target_col} by {col}")
        plt.xticks(rotation=45, ha="right")
        plt.tight_layout()
        plt.show()

# -------- 5) Outlier diagnostics (IQR method) --------
outlier_summary = []
for col in numeric_cols:
    s = df1[col].dropna()
    if s.empty:
        continue
    q1, q3 = s.quantile([0.25, 0.75])
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    outliers = ((df1[col] < lower) | (df1[col] > upper)).sum()
    outlier_summary.append({
        "column": col,
        "lower_bound": round(lower, 3),
        "upper_bound": round(upper, 3),
        "outlier_count": int(outliers),
        "outlier_pct": round(outliers / len(df1) * 100, 2)
    })

outlier_df = pd.DataFrame(outlier_summary).sort_values("outlier_pct", ascending=False)
print("\nOutlier summary (IQR):")
display(outlier_df)

# -------- 6) Multivariate segment snapshot --------
if numeric_cols:
    # simple behavior segmentation by quantiles of target-like column if available
    if target_col in numeric_cols:
        seg = df1.copy()
        valid_target = seg[target_col].dropna()
        unique_count = valid_target.nunique()

        if unique_count >= 2:
            requested_q = min(4, unique_count)
            _, bin_edges = pd.qcut(valid_target, q=requested_q, retbins=True, duplicates="drop")
            n_bins = len(bin_edges) - 1

            if n_bins >= 2:
                label_pool = ["Low", "Mid-Low", "Mid", "Mid-High", "High"]
                labels = label_pool[:n_bins] if n_bins <= len(label_pool) else [f"Q{i+1}" for i in range(n_bins)]

                seg.loc[valid_target.index, "target_band"] = pd.qcut(
                    valid_target,
                    q=requested_q,
                    labels=labels,
                    duplicates="drop"
                )
                seg["target_band"] = pd.Categorical(seg["target_band"], categories=labels, ordered=True)

                segment_profile = seg.groupby("target_band", observed=False)[numeric_cols].mean().round(2)
                print("\nSegment profile (means by target band):")
                display(segment_profile)
            else:
                print("\nSegment profile skipped: insufficient distinct quantile bins after duplicate-edge removal.")
        else:
            print("\nSegment profile skipped: target-like column has fewer than 2 unique non-null values.")

print("\nAdvanced EDA completed.")
```
